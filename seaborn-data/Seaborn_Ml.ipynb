## Imports

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Datasets

```python
mpg = sns.load_dataset('mpg')
```

```python
mpg.head()
```

```python
mpg.info()
```

## Distribution plot 

```python
sns.displot(mpg['mpg'], kde=True, bins=20)
plt.show()
```

## Scatter Plot

```python
mpg.head(30)
```

```python
plt.figure(figsize=(11, 9))
sns.scatterplot('mpg', 'horsepower', data=mpg)
plt.show()
```

```python
plt.figure(figsize=(11, 9))
sns.scatterplot(x='mpg', y='horsepower', data=mpg, hue='cylinders', palette='magma')
plt.show()
```

```python
mpg.head()
```

```python
plt.figure(figsize=(11,9))
sns.scatterplot(x='mpg', y='horsepower', data=mpg, hue='cylinders', size='weight')
plt.show()
```

## Joint Plot

```python
plt.figure(figsize=(11, 9))
sns.jointplot(x='mpg', y="horsepower", data=mpg)
plt.show()
```

```python
plt.figure(figsize=(11, 9))
sns.jointplot(x="weight", y='mpg', data=mpg, kind='hex')
plt.show()
```

```python
plt.figure(figsize=(11, 9))
sns.jointplot(x="weight", y='mpg', data=mpg, kind='reg')
plt.show()
```

## Pairplot
pairplot plots pairwaise relationship in an entire dataframe (Between Numerical Columns) and support a color hue arguement (for categorical columns)

```python
mpg.head()
```

```python
mpg_num = mpg[['mpg', 'cylinders', 'weight', 'horsepower']]
```

```python
mpg_num.head() 
```

```python
plt.figure(figsize=(15,10))
sns.pairplot(mpg_num)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.pairplot(mpg_num, hue='cylinders')
plt.show()
```

## Categorical Plots

```python
plt.figure(figsize=(11,9))
sns.boxplot(x='cylinders', y='mpg', data=mpg)
plt.show()
```

```python
plt.figure(figsize=(11,9))
sns.barplot(x='cylinders', y='mpg', data=mpg)
plt.show()
```

```python
plt.figure(figsize=(11,9))
sns.countplot(x='cylinders', data=mpg)
plt.xlim(2,10)
plt.show()
```

```python
mpg['cylinders'].value_counts()
```

### Heatmap

```python
mpg.corr()
```

```python
plt.figure(figsize=(11,9))
sns.heatmap(mpg.corr(), cmap='summer_r', annot=True)
plt.show()
```

### Facet Grids

```python
fg = sns.FacetGrid(mpg, col='cylinders', row='origin')
fg = fg.map(plt.scatter, 'mpg', 'weight' )
```

```python
fg = sns.FacetGrid(mpg, col='cylinders', row='origin', hue='horsepower')
fg = fg.map(plt.scatter, 'mpg', 'weight')
```

##  Regression Plots

```python
mpg.head(50)
```

```python
plt.figure(figsize=(11,9))
sns.lmplot(x='weight', y='mpg', data=mpg)
plt.show()
```

```python
plt.figure(figsize=(11,9))
sns.lmplot(x='weight', y='mpg', hue='origin', data=mpg, palette='magma')
plt.show()
```

```python
plt.figure(figsize=(11,9))
sns.lmplot(x='weight', y='mpg', col='origin', data=mpg, palette='magma')
plt.show()
```
